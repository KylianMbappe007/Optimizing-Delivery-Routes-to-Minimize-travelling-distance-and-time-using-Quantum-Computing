```python
import numpy as np
import matplotlib.pyplot as plt


L = 1.0  # deriving the length of the well
hbar = 1.0545718e-34  # [h / 2pie] = Reduced Planck's constant
m = 9.10938356e-31  # constant mass of an electron

# this simply denotes the boundary condition the boundary equation in an infinite potential well
def psi_n(x, n, L):
    return np.sqrt(2/L) * np.sin(n * np.pi * x / L)

# Defining the energy levels for the infinite potential well as the n value [number of waves]
#continue to increase DISCRETELY!!!!
def E_n(n, L):
    return (n**2 * np.pi**2 * hbar**2) / (2 * m * L**2)

# Plot the wavefunctions inside that boundary space, the HOLLOW SPACE BETWEEN THE TWO BOUNDARY DISTANCE
x = np.linspace(0, L, 1000)
n_values = [1, 2, 3]

plt.figure(figsize=(10, 6))
for n in n_values:
    plt.plot(x, psi_n(x, n, L), label=f'n={n}')

plt.title('Wavefunctions of an Infinite Potential Well')
plt.xlabel('Position (x)')
plt.ylabel('Wavefunction Ïˆ(x)')
plt.legend()
plt.grid(True)
plt.show()
```
